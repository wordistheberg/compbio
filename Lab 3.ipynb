```python
import sys
import operator

infile = open('romeoandjuliet.txt')
d = dict()
punctuation = '.,:;/][)(?!*'
for line in infile:
    line = line.strip()
    line = line.lower()
    words = line.split()
    for word in words:
        for i in punctuation:
            if word[-1] == i:
                word = word[:-1]
        if word in d:
            d[word] = d[word] + 1
        else:
            d[word] = 1
sorted_d = dict(sorted(d.items(), key=operator.itemgetter(1), reverse=True))
print(sorted_d)



```

```python
import matplotlib.pyplot as plt
%matplotlib inline

data = list(sorted_d.values())
words = list(sorted_d.keys())
line_out = plt.bar(height = data, x = words)
```

The 10 most frequent words from Romeo and Juliet are "and", "the", "i", "to", "a", "of", "my", "is", "that", and "in". 
For the least frequent, it's probably a couple thousand that are tied with only one use. This means that the words that are most frequent are truly just common in real life. Grammatically, words like "and" and "the" and "i" are needed to start/continue sentences, just as the other words are. The least frequent words, like "perchance" or "remnants" or "sovereign" have very specific uses that are necessary to the play, but not necessary to all/most sentences like the most frequent words are.

```python
import numpy as np
log_data = np.log10(data)
log_rank = np.log10(list(range(1,len(data) + 1)))

plt.bar(height = log_data, x = log_rank)
```

This graph looks like a steady exponential downward curve. I would interpret this to say that the higher rankings of the words have a significantly lower word count than the early rankings of the words.
