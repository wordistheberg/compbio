```python
from Bio import Entrez
from Bio import SeqIO
records = []
Entrez.email = 'jwberg@berkeley.edu'
handle = Entrez.esearch(db = 'nucleotide',
                        term = 'Human immunodeficiency virus 1 Envelope surface glycoprotein gp120',
                        sort ='relevance',
                       idtype='acc')
fastas = []
for i in Entrez.read(handle)['IdList']:
    handle = Entrez.efetch(db = 'nucleotide', id=i, rettype='fasta', retmode='text')
    for record in (SeqIO.parse(handle, "fasta")):
        fastas.append(record)
```

```python
SeqIO.write(fastas, 'fastas.fa', 'fasta')
```
